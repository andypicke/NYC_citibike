# This notebook contains EDA on citibike stations.
* Data on stations is stored in a database called **citibike_database.db3**

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import sqlite3
```

```python
# connect to database
con = sqlite3.connect('data/citibike_database.db3')
```

```python
df = pd.read_sql_query('SELECT * FROM stations',con)
df.head()
```

```python
import folium
map_osm = folium.Map(location=[40.73, -73.97],zoom_start=11.5)
map_osm
```

```python
# Add markers for stations
lat_list = df.lat.values#[0:10]
lon_list = df.lon.values#[0:10]
names = df.name#[0:10]

# add station locations:
for lat, lon, name in zip(lat_list, lon_list, names):
    #print(lat,lon,name)
#   folium.Marker([lat, lon], popup=name).add_to(map_osm)
    folium.CircleMarker([lat, lon], popup=name, radius=2, fill_opacity=0.7, fill_color='blue', color='blue' ).add_to(map_osm)
map_osm
```
