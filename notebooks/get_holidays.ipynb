# This notebook gets a list of US holidays, for use in NYC citibike analysis
* The goal is to predict the number of rides per day; holidays can have a large effect on this so I want to include them as a variable in my model.s

```python
import requests
import pandas as pd

year=2015
parameters = {'year':year,'key':'f25f83af-c733-43e2-850e-bc37a2fec702','country':'US'}
url = 'https://holidayapi.com/v1/holidays?'
response = requests.get(url, params=parameters);
b = response.json()['holidays']
#b.keys() # each date is a key
```

```python
df=pd.DataFrame()
for key in b.keys():
    #print(b[key])
    #df = pd.DataFrame.from_dict(b[key])
    df=df.append(pd.DataFrame.from_dict(b[key]),ignore_index=True)
```

## Get holidays for all years of citibike data
- Make sure to delete 'holiday' table before running, if exists

```python
import requests
import pandas as pd
import sqlite3

for year in [2014,2015,2016]:
    parameters = {'year':year,'key':'f25f83af-c733-43e2-850e-bc37a2fec702','country':'US'}
    url = 'https://holidayapi.com/v1/holidays?'
    response = requests.get(url, params=parameters);
    b = response.json()['holidays']
    #b.keys() # each date is a key
    df=pd.DataFrame()
    for key in b.keys():
        #print(b[key])
        #df = pd.DataFrame.from_dict(b[key])
        df=df.append(pd.DataFrame.from_dict(b[key]),ignore_index=True)
    
    df.date = pd.to_datetime(df.date)
    df.observed = pd.to_datetime(df.observed)
    df['yday'] = df['observed'].dt.dayofyear
    df['year'] = df['observed'].dt.year

    con = sqlite3.connect("data/citibike_database.db3")
    df.to_sql('holidays',con=con,if_exists='append',index=False)
```
